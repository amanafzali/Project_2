```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import reuters, stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.util import ngrams
from nltk.stem import WordNetLemmatizer
import re
from collections import Counter
from wordcloud import WordCloud

%matplotlib inline

# Code to download corpora
import nltk
nltk.download('stopwords')
nltk.download('reuters')
nltk.download('punkt')

import warnings
warnings.filterwarnings('ignore')
```

```python
# Set path to CSV and read in CSV
csv_path = Path('bitcoin_tweets_3.csv')
tweets_df = pd.read_csv(csv_path)
tweets_df.set_index(pd.to_datetime(tweets_df['date'], infer_datetime_format=True), inplace=True)
tweets_df.drop(columns=['date'], inplace=True)
tweets_df.head()
```

```python
# Create Corpus
corpus = tweets_df['text']
```

```python
# Lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# Define preprocess function
def process_text(doc):
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', doc)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw]
    return output
```

```python
# Define the counter function
def word_counter(corpus): 
    # Combine all articles in corpus into one large string
    big_string = " ".join(corpus)
    processed = process_text(big_string)
    top_15 = dict(Counter(processed).most_common(15))
    return pd.DataFrame(list(top_15.items()), columns=['word', 'count'])
```

```python
word_counter(corpus)
```

```python
# Function for Bigram Counter
def bigram_counter(corpus): 
    #Combine all articles in corpus into one large string
    big_string = ' '.join(corpus)
    processed = process_text(big_string)
    bigrams = ngrams(processed, n=2)
    top_10 = dict(Counter(bigrams).most_common(10))
    return pd.DataFrame(list(top_10.items()), columns=['bigram', 'count'])
```

```python
# Display & Count Bigram's
bigram_counter(corpus)
```

```python
# Function to Process text 
def process_text(doc):
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', doc)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw]
    return ' '.join(output)
```

```python
# Reformat text into string
big_string = ' '.join(corpus)
input_words = process_text(big_string)
```

```python
def wordcloud(text, title=""):
    df_cloud = WordCloud(width=500, colormap='RdYlBu').generate(input_words)
    plt.imshow(df_cloud)
    plt.axis("off")
    fontdict = {"fontsize": 48, "fontweight" : "bold"}
    plt.title(title, fontdict=fontdict)
    plt.show()
```

```python
# Generate the Bitcoin word cloud
wordcloud(tweets_df.text.str.cat(), title="Bitcoin Tweets Word Cloud")
```

```python
# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Sentiment calculation based on compound score
def get_sentiment(score):
    """
    Calculates the sentiment based on the compound score.
    """
    result = 0
    if score >= 0.05:
        result = 1
    elif score <= -0.05:
        result = -1
       
    return result
```

```python
text = {
    'tweet_compound': [],
    'tweet_pos': [],
    'tweet_neu': [],
    'tweet_neg': [],
    'tweet_sent':[],
}
```

```python
# For loop to populate DF W/Sentiment Scores
for index, row in tweets_df.iterrows():
    try:
        tweet_sentiment = analyzer.polarity_scores(row['text'])
        text['tweet_compound'].append(tweet_sentiment['compound'])
        text['tweet_pos'].append(tweet_sentiment['pos'])
        text['tweet_neu'].append(tweet_sentiment['neu'])
        text['tweet_neg'].append(tweet_sentiment['neg'])
        text['tweet_sent'].append(get_sentiment(tweet_sentiment['compound']))
        
        
    except AttributeError:
        pass
    
tweet_sentiment_df = pd.DataFrame(text)

tweet_sentiment_df.head(20)
```
